```python
import sympy
from einsteinpy.symbolic import MetricTensor, RiemannCurvatureTensor, RicciTensor, RicciScalar, ChristoffelSymbols
from sympy import exp

sympy.init_printing() 

#defining the symbols 
syms = sympy.symbols("t r theta phi")
c = sympy.symbols("c")

# defining the metric
metric = [[0 for i in range(4)] for i in range(4)]
metric[0][0] = -1
metric[1][1] = syms[1]
metric[2][2] = syms[1]**2
metric[3][3] = (syms[1]**2)*(sympy.sin(syms[2])**2)
# creating metric object
metric_obj = MetricTensor(metric, syms)

#The metric tensor
metric_obj.tensor()

ch = ChristoffelSymbols.from_metric(metric_obj)

rm1 = RiemannCurvatureTensor.from_christoffels(ch)
rm1.tensor()
```
